```python
df.groupby('category')['quantity'].sum()
```

```python
plt.figure(figsize=(20,5))
ax=sns.barplot(x=list(df.groupby('category').sum().sort_values(by='total',ascending=False).index),
               y=list(df.groupby('category').sum().sort_values(by='total',ascending=False).total))
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
plt.xlabel('Category',fontsize=20, color='red')
plt.ylabel('Total', fontsize=20, color='red')
plt.title('Revenue of each category', fontsize=25, color='red')
plt.show()
print(df.groupby('category').sum()['total'].sort_values(ascending=False).head(5))
```

```python


#Highest quantity
plt.figure(figsize=(20,5))
ax=sns.barplot(x=list(df.groupby('category').sum().sort_values(by='quantity',ascending=False).index),
               y=list(df.groupby('category').sum().sort_values(by='quantity',ascending=False).quantity))
ax.set_xticklabels(ax.get_xticklabels(),rotation=40)
plt.xlabel('Category',fontsize=20, color='red')
plt.ylabel('Quantity', fontsize=20, color='red')
plt.title('Total quantity of each category', fontsize=25, color='red')
plt.show()
print(df.groupby('category').sum()['quantity'].sort_values(ascending=False).head(10))

```
